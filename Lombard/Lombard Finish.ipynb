```python
sql_query = "SELECT a.[PartnerID] \
      ,IIF([Sex]='M',1,0) SEX \
      ,[Age] \
      ,[CreditCondition] \
      ,DateDiff(Day,[LombardFirstDealDate],GetDATE()) LifeTime \
      ,DateDiff(Day,[IssueDate],GetDATE()) DayFromLast \
      ,[BudetPerezalog] \
      ,[SumAmountzb] \
      ,[AVGAmountzb] \
      ,[ZB] \
      ,[PercentSMS] \
      ,CAST(c.QntPromo as Float)/ZB  PercentPromo \
      ,[PercentBudetPerezalog] \
      ,[PercentGood] \
      ,[PercentNew] \
      ,[PercentPerez] \
      ,[PercentOSN] \
      ,[PercentBuyUp] \
      ,[AVGDneyZaloga] \
      ,[AVGDneySProshZB] \
      ,[AVGWeightGross] \
      ,[AVGCostPerGram] \
      ,[AVGCost] \
      ,[MaxCost] \
      ,[AVGPercentPerDay] \
      ,[MINPercentPerDay] \
      ,[MAXPercentPerDay] \
      ,[isGoodZB] \
      ,[BudetSledVisit3M] \
      ,[MoreGood] \
  FROM [Gold585].[dbo].[LombardCommonSMS] a \
  Left join LombardPromotion c on a.Code = c.Code and a.PartnerID = c.PartnerID \
  Where a.IssueDate >= convert(datetime,'2016-06-01') and a.IssueDate <= convert(datetime,'2018-06-01') and a.Category in ('Новая cсуда', 'Перезалог') \
  AND  a.IssueDateTime = (Select MAX(IssueDateTime) From [Gold585].[dbo].[LombardCommonSMS] b where a.PartnerID = b.PartnerID \
  and b.Category in ('Новая cсуда', 'Перезалог') and b.IssueDate <= convert(datetime,'2018-06-01') \
  group by b.PartnerID) \
  Order By a.IssueDateTime"
data = pd.read_sql(sql_query,conn)
```

```python
print(data.shape)
print(data[data['ZB']==1].shape)
print(np.percentile(data['ZB'],99)) # у 95 процентов посетителей < 73 ЗБ
```

```python
df = data[(data['ZB']<np.percentile(data['ZB'],99)) ].fillna(0)
df.reset_index(drop=True,inplace = True)
target=df.drop(['PartnerID','BudetPerezalog','DayFromLast','AVGCost', 'MaxCost','isGoodZB','BudetSledVisit3M','MINPercentPerDay', 'MAXPercentPerDay', ], axis=1)
target = pd.get_dummies(target, columns = ['SEX','CreditCondition'])
X = StandardScaler().fit_transform(target)

from sklearn.decomposition import PCA
# PCA
pca = PCA(n_components=2) #HOW MANY AXIS DO WE WANT
pca.fit_transform(X)
targert2 = pca.transform(X)


fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 18
fig_size[1] = 12

plt.scatter(targert2[:, 0], targert2[:, 1])
```

```python
kmeans = KMeans(n_clusters=9, random_state=42,n_jobs=-2).fit(X) ## нормальные random_state: 4
labels = kmeans.predict(X)
df['labels']=pd.DataFrame(labels)

targert2 = pd.DataFrame(targert2,columns=('X','Y'))
targert2['L'] = pd.DataFrame(labels)

fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 18
fig_size[1] = 12

plt.scatter(targert2['X'], targert2['Y'],c =labels)
```
